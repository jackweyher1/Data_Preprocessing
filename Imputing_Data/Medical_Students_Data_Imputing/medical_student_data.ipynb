# Manual Pandas Data Imputing

```python
# Get  Columns with Null values and split by datatype

for col in data.columns[data.isna().any()]:
    print(f'{col}: {data[col].dtypes}')
```

```python
# Replace numeric column with null values with the mean of that feature
data_copy = data.copy()

num_cols = ['Student ID', 'Age', 'Height', 'Weight', 'BMI', 'Temperature', 'Heart Rate', 'Blood Pressure', 'Cholesterol']

for col in num_cols:
    data_copy[col] = data_copy[col].fillna(data_copy[col].mean())
```

```python
data.info()
```

```python
data_copy.info()
```

```python
cat_cols = ['Gender', 'Blood Type', 'Diabetes', 'Smoking']

for col in cat_cols:
    data_copy[col] = data_copy[col].fillna(data_copy[col].mode()[0])
```

```python
data.info()
```

```python
data_copy.info()
```

# Fill null values using scikit learn Imputer method

```python
from sklearn.impute import SimpleImputer
import numpy as np

data_copy_2 = data.copy()

# Create a imputer object for numeric columns
# strategy tells the imputer how to fill null values (mean for numeric)
# missing_values tells the imputer what values to fill (np.nan / null values)
num_imputer = SimpleImputer(strategy='mean', missing_values=np.nan) 

for col in num_cols:
    num_imputer.fit(data_copy_2[[col]])
    data_copy_2[col] = num_imputer.transform(data_copy_2[[col]])
```

```python
data_copy_2.info()
```

```python
# Create a string Imputer object

# strategy for this string Imputer is most_frequent
# missing_values is still np.nan values
str_imputer = SimpleImputer(strategy='most_frequent', missing_values=np.nan)

for col in cat_cols:
    str_imputer.fit(data_copy_2[[col]])

    # For string/categorical columns, we need to use ravel() to bring back to a vector (dont need to for num cols)
    data_copy_2[col] = str_imputer.transform(data_copy_2[[col]]).ravel()
```
