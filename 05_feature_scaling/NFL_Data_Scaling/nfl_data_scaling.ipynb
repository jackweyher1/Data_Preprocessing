# Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import random
```

# Create NFL Dataset

Note: Numpy's random module excludes the high end, for example random.randint(14, 17) only returns 14 15 and 16.

```python
num_players     = 50
position        = ['Running Back' for _ in range(num_players)]
player_id       = random.sample(range(1, 1000), num_players)
rushing_yards   = np.random.normal(1200, 200, num_players).clip(600, 1800).round(0)
receiving_yards = np.random.normal(600, 150, num_players).clip(100, 1100).round(0)
games_played    = np.random.randint(14, 18, num_players)
touches         = np.random.normal(240, 40, num_players).clip(40, 340).round(0)
fumbles         = np.random.randint(0, 5, num_players)
nfl_df = pd.DataFrame(
    {
        'Position': position,
        'Player ID': player_id,
        'Rushing Yards': rushing_yards,
        'Receiving Yards': receiving_yards,
        'Games Played': games_played,
        'Touches': touches,
        'Fumbles': fumbles
    }
)
```

# Quick EDA

```python
nfl_df.head()
```

# Normalization (Min-Max Sclaing)
- Scales values to a fixed range, usually [0, 1]
- Best for: when you need bounded values, neural networks, KNN
- Sensitive to outliers

```python
from sklearn.preprocessing import MinMaxScaler

norm_scaler = MinMaxScaler()

cols_to_scale = ['Rushing Yards', 'Receiving Yards', 'Games Played', 'Touches', 'Fumbles']

nfl_df_norm = nfl_df.copy()
nfl_df_norm[cols_to_scale] = norm_scaler.fit_transform(nfl_df_norm[cols_to_scale])
```

```python
nfl_df_norm.head()
```

# Standardization (Z-Scaling)
- Centers data around mean=0 with std=1
- Best for: most ML algorithms, when data is normally distributed
- Less sensitive to outliers than MinMax



```python
from sklearn.preprocessing import StandardScaler

stand_scaler = StandardScaler()

nfl_df_stand = nfl_df.copy()
nfl_df_stand[cols_to_scale] = stand_scaler.fit_transform(nfl_df_stand[cols_to_scale])
```

```python
nfl_df_stand.head()
```

Deciding which type of scaling we use depends on the application, but typicall Standardization is the better choice because it keeps the same distribution as the unscaled dataset.

# Robust Scaling (Mediam & IQR Scaling)
- Uses median and interquartile range (IQR) instead of mean/std
- Best for: data with significant outliers
- Most outlier-resistant of the common scalers

```python
from sklearn.preprocessing import RobustScaler

robust_scaler = RobustScaler()

nfl_df_robust = nfl_df.copy()
nfl_df_robust[cols_to_scale] = robust_scaler.fit_transform(nfl_df_robust[cols_to_scale])
```

```python
nfl_df_robust.head()
```

# Max Abs Scaling (Absolute Value Scaler)
- Scales by dividing by the maximum absolute value, range [-1, 1]
- Best for: sparse data, already centered at zero

```python
from sklearn.preprocessing import MaxAbsScaler

abs_scaler = MaxAbsScaler()

nfl_df_abs = nfl_df.copy()

nfl_df_abs[cols_to_scale] = abs_scaler.fit_transform(nfl_df_abs[cols_to_scale])
```

```python
nfl_df_abs.head()
```

# Log Transformer (NumPy Log Scaling) 
- Reduces skewness in heavily skewed data
- Best for: right-skewed distributions, exponential data

```python
nfl_df_log = nfl_df.copy()

nfl_df_log[cols_to_scale] = np.log1p(nfl_df_log[cols_to_scale])
```

```python
nfl_df_log.head()
```

# Power Transformer
- Makes data more Gaussian-like
- Best for: when normality is required

```python
nfl_df.describe().T
```

```python
from sklearn.preprocessing import PowerTransformer

# Two methods
pt_yj = PowerTransformer(method='yeo-johnson')  # handles negative values

nfl_df_pt_yj = nfl_df.copy()

# Since our data has zeroes in the fumbles feature, we hae to use the yeo johnson method
nfl_df_pt_yj[cols_to_scale] = pt_yj.fit_transform(nfl_df_pt_yj[cols_to_scale])
```

```python
nfl_df_pt_yj.head()
```

```python
# Make a small shift of 1 for the Fumbles column to be able to use box-cox 
pt_bc = PowerTransformer(method='box-cox')

nfl_df_pt_bc = nfl_df.copy()
nfl_df_pt_bc['Fumbles'] = nfl_df_pt_bc['Fumbles'] + 1 # Shift 1 to get rid of zeroes

nfl_df_pt_bc[cols_to_scale] = pt_bc.fit_transform(nfl_df_pt_bc[cols_to_scale])
```

```python
nfl_df_pt_bc.head()
```

# Quantile Transformer (Uniform or Normal scaler)
- Maps data to a uniform or normal distribution
- Best for: non-linear scaling, when you want to force a specific distribution
- Most aggressive transformation

```python
from sklearn.preprocessing import QuantileTransformer

# Uniform Scale
qt_uni = QuantileTransformer(output_distribution='uniform')

nfl_df_qt_uni = nfl_df.copy()
nfl_df_qt_uni[cols_to_scale] = qt_uni.fit_transform(nfl_df_qt_uni[cols_to_scale])
```

```python
nfl_df_qt_uni.head()
```

```python
qt_norm = QuantileTransformer(output_distribution='normal')

nfl_df_qt_norm = nfl_df.copy()
nfl_df_qt_norm[cols_to_scale] = qt_norm.fit_transform(nfl_df_qt_norm[cols_to_scale])
```

```python
nfl_df_qt_norm.head()
```
