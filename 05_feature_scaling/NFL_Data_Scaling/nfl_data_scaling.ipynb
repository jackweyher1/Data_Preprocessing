# Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import random
```

# Create NFL Dataset

```python
num_players     = 50
position        = ['Running Back' for _ in range(num_players)]
player_id       = random.sample(range(1, 1000), num_players)
rushing_yards   = np.random.normal(1200, 200, num_players).clip(600, 1800).round(0)
receiving_yards = np.random.normal(600, 150, num_players).clip(100, 1100).round(0)
games_played    = np.random.randint(14, 17, num_players)
touches         = np.random.normal(240, 40, num_players).clip(40, 340).round(0)
fumbles         = np.random.randint(0, 4, num_players)
nfl_df = pd.DataFrame(
    {
        'Position': position,
        'Player ID': player_id,
        'Rushing Yards': rushing_yards,
        'Receiving Yards': receiving_yards,
        'Games Played': games_played,
        'Touches': touches,
        'Fumbles': fumbles
    }
)
```
