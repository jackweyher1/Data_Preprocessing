# Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import random
```

# Create NFL Dataset

Note: Numpy's random module excludes the high end, for example random.randint(14, 17) only returns 14 15 and 16.

```python
num_players     = 50
position        = ['Running Back' for _ in range(num_players)]
player_id       = random.sample(range(1, 1000), num_players)
rushing_yards   = np.random.normal(1200, 200, num_players).clip(600, 1800).round(0)
receiving_yards = np.random.normal(600, 150, num_players).clip(100, 1100).round(0)
games_played    = np.random.randint(14, 18, num_players)
touches         = np.random.normal(240, 40, num_players).clip(40, 340).round(0)
fumbles         = np.random.randint(0, 5, num_players)
nfl_df = pd.DataFrame(
    {
        'Position': position,
        'Player ID': player_id,
        'Rushing Yards': rushing_yards,
        'Receiving Yards': receiving_yards,
        'Games Played': games_played,
        'Touches': touches,
        'Fumbles': fumbles
    }
)
```

```python
nfl_data_norm = nfl_df.copy()
nfl_data_stand = nfl_df.copy()
```

# Quick EDA

```python
nfl_df.head()
```

# Normalization Sclaing

```python
from sklearn.preprocessing import MinMaxScaler

norm_scaler = MinMaxScaler()

cols_to_scale = ['Rushing Yards', 'Receiving Yards', 'Games Played', 'Touches', 'Fumbles']

nfl_data_norm[cols_to_scale] = norm_scaler.fit_transform(nfl_data_norm[cols_to_scale])
```

```python
nfl_data_norm.head()
```

# Standardization Scaling

```python
from sklearn.preprocessing import StandardScaler

stand_scaler = StandardScaler()

nfl_data_stand[cols_to_scale] = stand_scaler.fit_transform(nfl_data_stand[cols_to_scale])
```

```python
nfl_data_stand.head()
```

Deciding which type of scaling we use depends on the application, but typicall Standardization is the better choice because it keeps the same distribution as the unscaled dataset.
